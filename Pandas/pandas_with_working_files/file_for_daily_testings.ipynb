```python
import pandas as pd
file_serbia = pd.read_excel('Daily Test Report(17.01.2023).xlsx')
file_serbia.head()
```

```python
#Print all info about date what you need
need_date = '20230117'
file_for_needed_date = file_serbia[ file_serbia.Date == need_date ]
file_for_needed_date
```

```python
import matplotlib.pyplot as plt
#count Delivered messages, not sent and not Delivered
data = file_for_needed_date.groupby('Status of delivery').count()['Date']
# Status of delivery
# Delivered        41
# Not Delivered     2
# Not Sent          3
# Name: Date, dtype: int64

# build a chart for this value
generated = data['Delivered'] + data['Not Delivered'] + data['Not Sent']
data = data.append(pd.Series(generated, index = ['Generated']))
data = data.sort_values(ascending=False)
print(data)
data.plot.bar()
plt.show()
```

```python
import matplotlib.pyplot as plt
#count grayroute messages in some period of dates
telco_test = pd.read_excel('Report_telemach_2022-11-28.xlsx', sheet_name='Illegal Traffic')
need_date = 1
file_with_gray = telco_test[ telco_test.Date >= need_date ]
data = file_with_gray.groupby('Fraud Type').count()['Date']
print(data)

data.plot.bar()
plt.show()
```

```python
#create Dataframe with all grayroute for sending mail to MNO
telco_test = pd.read_excel('Report_telemach_2022-11-28.xlsx', sheet_name='Illegal Traffic')
date = 28
month = 11
year = 2022
file_with_gray = telco_test[ telco_test.Date == need_date ]
file_with_gray = file_with_gray.drop(columns=['operator','celltest_type', 'alert_name','source'])
file_with_gray.to_excel(f"Gray_messages_{date}.{month}.{year}.xlsx", index=False)

telco_test = pd.read_excel('Report_telemach_2022-11-28.xlsx', sheet_name='Flash_Calls')
file_with_flashcalls = telco_test[ telco_test.Day == need_date ]
file_with_flashcalls = file_with_flashcalls.drop(columns=['Service', 'operator', 'alert_name', 'source'])
file_with_flashcalls.rename(columns=
{
    'Calling Number' : 'Caller',
    'Caller Number' : 'DA',
    "Call Date/Time" : 'Timestamp'
},inplace=True)
file_with_flashcalls.to_excel(f"Flashcalls_{date}.{month}.{year}.xlsx", index=False)
```
