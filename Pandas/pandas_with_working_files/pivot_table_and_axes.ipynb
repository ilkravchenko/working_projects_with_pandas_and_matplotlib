```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_excel('Report_telemach_2022-11-28.xlsx', sheet_name='Illegal Traffic')
#data

pivot_table_to_plot = np.round(pd.pivot_table(data, values ='alert_name', index=['Date'],
                            columns = ['Fraud Type'], aggfunc=np.count_nonzero, fill_value=0),2)
pivot_table_to_print = np.round(pd.pivot_table(data, values ='alert_name', index=['Fraud Type'],
                            columns = ['Date'], aggfunc=np.count_nonzero, fill_value=0),2)

print(pivot_table_to_print.to_string())

fig, ax = plt.subplots(figsize=(20,5))

ax.plot(pivot_table_to_plot, marker='o', label=['Illegal GT', 'Sim-boxes','Spoofing'])

ax.set_title('Detected Fraud', fontsize=20, font='arial')
ax.set_xlabel("Date of November",fontsize=14, font='arial')
ax.set_ylabel("Count of Fraud",fontsize=14, font='arial')
ax.grid(color = 'black', linestyle='--',alpha = 0.5)
ax.legend(loc='best')
ax.set_xticks(np.arange(1, 29, 1))
ax.set_yticks(np.arange(0, 10, 1))
```
